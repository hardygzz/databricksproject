```python
dbutils.widgets.removeAll()

dbutils.widgets.text("storageName", "adlsbsriskdev")
dbutils.widgets.text("credentialName", "cred-sbsrisk-dev")
dbutils.widgets.text("catalogName", "sbsrisk_dev")
```

```python
%sql
SELECT
  '${storageName}' AS storageName,
  '${credentialName}' AS credentialName,
  '${catalogName}' AS catalogName;
```

```python
%sql
CREATE EXTERNAL LOCATION IF NOT EXISTS `extl_raw`
URL 'abfss://raw@${storageName}.dfs.core.windows.net/'
WITH (STORAGE CREDENTIAL `${credentialName}`);

CREATE EXTERNAL LOCATION IF NOT EXISTS `extl_bronze`
URL 'abfss://bronze@${storageName}.dfs.core.windows.net/'
WITH (STORAGE CREDENTIAL `${credentialName}`);

CREATE EXTERNAL LOCATION IF NOT EXISTS `extl_silver`
URL 'abfss://silver@${storageName}.dfs.core.windows.net/'
WITH (STORAGE CREDENTIAL `${credentialName}`);

CREATE EXTERNAL LOCATION IF NOT EXISTS `extl_gold`
URL 'abfss://gold@${storageName}.dfs.core.windows.net/'
WITH (STORAGE CREDENTIAL `${credentialName}`);

CREATE EXTERNAL LOCATION IF NOT EXISTS `extl_metastore`
URL 'abfss://metastore@${storageName}.dfs.core.windows.net/'
WITH (STORAGE CREDENTIAL `${credentialName}`);
```

```python
%sql
CREATE CATALOG IF NOT EXISTS `${catalogName}`
MANAGED LOCATION 'abfss://metastore@${storageName}.dfs.core.windows.net/${catalogName}/';

CREATE SCHEMA IF NOT EXISTS `${catalogName}`.bronze;
CREATE SCHEMA IF NOT EXISTS `${catalogName}`.silver;
CREATE SCHEMA IF NOT EXISTS `${catalogName}`.gold;
```

```python
%sql
SHOW STORAGE CREDENTIALS;
SHOW EXTERNAL LOCATIONS;

SHOW CATALOGS;
SHOW SCHEMAS IN `${catalogName}`;

DESCRIBE EXTERNAL LOCATION extl_raw;
DESCRIBE EXTERNAL LOCATION extl_bronze;
DESCRIBE EXTERNAL LOCATION extl_silver;
DESCRIBE EXTERNAL LOCATION extl_gold;
DESCRIBE EXTERNAL LOCATION extl_metastore;
```
